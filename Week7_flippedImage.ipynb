**Week 7 Homework Flipped Image**

```
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
url = "https://github.com/imageio/imageio-binaries/raw/master/images/chelsea.png"

#img = imageio.imread(url)
img = mpimg.imread(url)

img2 = np.flipud(img)
plt.imshow(img2)

plt.savefig("chelsea_fliped_vertically.png", dpi=200)
plt.show("chelsea_fliped_vertically.png")
print(url)
```
